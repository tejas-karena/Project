```python
data = data.astype({'Followers':'int','Authentic Engagement':'int','Engagement Avg.':'int'})
data.head()
```

```python
top20Followers = data.nlargest(20,'Followers')
top20Followers
```

```python
from matplotlib import pyplot as plt 
```

```python
plt.figure(figsize=(12,8))
fig2=plt.bar(top20Followers.Category,top20Followers.Followers,label='Folloers in Millions',width=0.8,color='orange')
plt.xticks(fontsize=12,rotation=90)
plt.ylabel("Followers in Millions",fontsize=16)
plt.xlabel("Category",fontsize=16)
plt.title("Top Twenty Countries by followers",fontsize=20)
plt.tight_layout()
plt.ticklabel_format(axis="y",style="plain")
```

```python
# Piechart
types=data.Category.value_counts()
types.reset_index()
```

```python
plt.pie(types.values,labels=types.index,labeldistance=1.2,autopct='%1.2f%%',shadow=True,startangle=60)
plt.title('Different categories',fontsize=20)
plt.plot()
```
